```python
coeff = [r,k,q,D,c,A,a]
```

```python
coeffs = pd.DataFrame(coeff)
coeffs.index=['r','k','q','D','c','A','a']
coeffs = coeffs.loc[:,coeffs.notna().all(axis=0)]
coeffs = coeffs.T
coeffs
```

```python
coeffs['r'] = np.random.permutation(coeffs['r'])
coeffs['k'] = np.random.permutation(coeffs['k'])
coeffs['q'] = np.random.permutation(coeffs['q'])
coeffs['D'] = np.random.permutation(coeffs['D'])
coeffs['c'] = np.random.permutation(coeffs['c'])
coeffs['A'] = np.random.permutation(coeffs['A'])
coeffs['a'] = np.random.permutation(coeffs['a'])
```

```python
coeffs = coeffs.loc[coeffs['A']*coeffs['q'] > coeffs['a']*coeffs['c']]
coeffs = coeffs.reset_index(drop=True)
coeffs
```

```python
X=[]
for i in range(coeffs.shape[0]):
  poly = [coeffs.iloc[i]['r']*coeffs.iloc[i]['q'],-coeffs.iloc[i]['r']*coeffs.iloc[i]['q']*coeffs.iloc[i]['k'],coeffs.iloc[i]['k']*(coeffs.iloc[i]['A']*coeffs.iloc[i]['q']-coeffs.iloc[i]['a']*coeffs.iloc[i]['c']),-coeffs.iloc[i]['a']*coeffs.iloc[i]['k']*coeffs.iloc[i]['c']*coeffs.iloc[i]['D']]
  root=np.roots(poly)
  X.append(root)
```

Main Theorem
The interior equilibrium point (*x*,*y*,*z*) is locally asymptotically stable if one the following

*  If DG>0, L1>0, L3>0 and L1L2>L3
*  If DG<0, L1>=0, L2>=0, L3>0 for beta<(2/3)
*  If DG<0, L1>0, L2>0, L1L2 = L3, then |arg(L)|>beta*pi/2, for all beta in[0,1)

```python
YZ=[]
for i in range(XXX.shape[0]):
  x1 = XXX.iloc[i]['x1']
  x2 = XXX.iloc[i]['x2']
  x3 = XXX.iloc[i]['x3']
  r = XXX.iloc[i]['r']
  k = XXX.iloc[i]['k']
  q = XXX.iloc[i]['q']
  D = XXX.iloc[i]['D']
  c = XXX.iloc[i]['c']
  A = XXX.iloc[i]['A']
  a = XXX.iloc[i]['a']
  y1=(r/q)*(1-x1/k)*(D+x1)
  y2=(r/q)*(1-x2/k)*(D+x2)
  y3=(r/q)*(1-x3/k)*(D+x3)
  z1=(c/q)*((D+x1)/(x1))
  z2=(c/q)*((D+x2)/(x2))
  z3=(c/q)*((D+x3)/(x3))
  yz=[r,k,q,D,c,A,a,x1,y1,z1,x2,y2,z2,x3,y3,z3]
  YZ.append(yz)
```

```python
XYZ = pd.DataFrame(YZ)
XYZ.columns = (['r','k','q','D','c','A','a','x1','y1','z1','x2','y2','z2','x3','y3','z3'])
XYZ
```

```python
L=[]
for i in range(XYZ.shape[0]):
  x1 = XYZ.iloc[i]['x1']
  x2 = XYZ.iloc[i]['x2']
  x3 = XYZ.iloc[i]['x3']
  y1 = XYZ.iloc[i]['y1']
  y2 = XYZ.iloc[i]['y2']
  y3 = XYZ.iloc[i]['y3']
  z1 = XYZ.iloc[i]['z1']
  z2 = XYZ.iloc[i]['z2']
  z3 = XYZ.iloc[i]['z3']
  r = XYZ.iloc[i]['r']
  k = XYZ.iloc[i]['k']
  q = XYZ.iloc[i]['q']
  D = XYZ.iloc[i]['D']
  c = XYZ.iloc[i]['c']
  A = XYZ.iloc[i]['A']
  a = XYZ.iloc[i]['a']
  l1_1 = a*z1 + 2*(r/k)*x1 +(q*D*y1)/((x1+D)**2) - r
  l2_1 = (q*q*D*x1*y1*z1)/((x1+D)**3) + A*c - a*c*z1 + 2*(r*a/k)*x1*z1 + (q*a*D*y1*z1)/((x1+D)**2) - r*a*z1
  l3_1 = (a*c*z1 - A*c)*(r - 2*(r/k)*x1 - q*D*y1/((x1+D)**2)) + (2*a*z1 - A)*(q*q*D*x1*y1*z1)/((x1+D)**3)

  l1_2 = a*z2 + 2*(r/k)*x2 +(q*D*y2)/((x2+D)**2) - r
  l2_2 = (q*q*D*x2*y2*z2)/((x2+D)**3) + A*c - a*c*z2 + 2*(r*a/k)*x2*z2 + (q*a*D*y2*z2)/((x2+D)**2) - r*a*z2
  l3_2 = (a*c*z2 - A*c)*(r - 2*(r/k)*x2 - q*D*y2/((x2+D)**2)) + (2*a*z2 - A)*(q*q*D*x2*y2*z2)/((x2+D)**3)

  l1_3 = a*z3 + 2*(r/k)*x3 +(q*D*y3)/((x3+D)**2) - r
  l2_3 = (q*q*D*x3*y3*z3)/((x3+D)**3) + A*c - a*c*z3 + 2*(r*a/k)*x3*z3 + (q*a*D*y3*z3)/((x3+D)**2) - r*a*z3
  l3_3 = (a*c*z3 - A*c)*(r - 2*(r/k)*x3 - q*D*y3/((x3+D)**2)) + (2*a*z3 - A)*(q*q*D*x3*y3*z3)/((x3+D)**3)

  L1L2L3 = [l1_1,l2_1,l3_1,l1_2,l2_2,l3_2,l1_3,l2_3,l3_3]

  L.append(L1L2L3)
```
